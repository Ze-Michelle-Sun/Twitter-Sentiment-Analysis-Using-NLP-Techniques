# Pre-processing Tweets

```python
def preprocess_tweet_text(tweet):
    tweet = tweet.lower()
    # Remove urls
    tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE)
    # Decode to ASCII and replace â€™ with '
    tweet = tweet.encode('ascii', 'ignore').decode()
    tweet = tweet.replace("â€™", "'")
    # Remove user @ references and '#' from tweet
    tweet = re.sub(r'\@\w+|\#','', tweet)
    # Remove punctuations
    tweet = tweet.translate(str.maketrans('', '', string.punctuation))
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tweet_tokens = word_tokenize(tweet)
    filtered_words = [w for w in tweet_tokens if not w in stop_words]
    
    ps = PorterStemmer()
    stemmed_words = [ps.stem(w) for w in filtered_words]
    #lemmatizer = WordNetLemmatizer()
    #lemma_words = [lemmatizer.lemmatize(w, pos='a') for w in stemmed_words]
    
    return " ".join(filtered_words)
```

# Sentiment analysis with TextBlob

https://textblob.readthedocs.io/en/dev/quickstart.html#sentiment-analysis

> - The sentiment property returns a named tuple of the form Sentiment(polarity, subjectivity). 
> -The polarity score is a float within the range [-1.0, 1.0]. 
> -The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

## Function for labelling tweets based on classification criteria described above.

```python
# define a function to get TextBlob sentiment analysis results
def get_textblob_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    sentiment = 'negative' if polarity < 0 else 'neutral' if polarity == 0 else 'positive'
    return polarity, subjectivity, sentiment

# apply the function to each row in the dataset
df_BGclean[['blob_polarity', 'blob_subjectivity', 'blob_sentiment']] = df_BGclean['Tweet'].apply(get_textblob_sentiment).apply(pd.Series)
df_EMclean[['blob_polarity', 'blob_subjectivity', 'blob_sentiment']] = df_EMclean['Tweet'].apply(get_textblob_sentiment).apply(pd.Series)
```

## Vader is a pre-trained model
> - Vader outputs something like this: {'neg': 0.0, 'neu': 0.436, 'pos': 0.564, 'compound': 0.3802}
> - Negative, neutral and positive are scores between 0 and 1
> - The compound value reflects the overall sentiment of the text. It's computed based on the values of negative, neutral and positive. It ranges from -1 (maximum negativity) to 1 (maximum positivity).

# Observations from the distribution plot of sentiment produced by TextBlob and VADER 
We noticed that sentiment produced by TextBlob and VADER tend to:

* VADER shows the  bimodal distribution, while TextBlob shows just unimodal.

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


# Define a function to compare the sentiment scores of TextBlob and VADER
def compare_sentiment(tweet):
    # Calculate the TextBlob polarity and subjectivity scores
    blob_polarity = TextBlob(tweet).sentiment.polarity
    blob_subjectivity = TextBlob(tweet).sentiment.subjectivity
    if blob_polarity > 0:
        blob_sentiment = 'positive'
    elif blob_polarity < 0:
        blob_sentiment = 'negative'
    else:
        blob_sentiment = 'neutral'
    # Calculate the VADER sentiment scores
    analyzer = SentimentIntensityAnalyzer()
    vader_scores = analyzer.polarity_scores(tweet)
    vader_compound = vader_scores['compound']
    if vader_compound >= 0.05:
        vader_sentiment = 'positive'
    elif vader_compound <= -0.05:
        vader_sentiment = 'negative'
    else:
        vader_sentiment = 'neutral'
    # Compare the sentiment scores of TextBlob and VADER
    if blob_sentiment == vader_sentiment:
        sentiment_comparison = 'agree'
    else:
        sentiment_comparison = 'disagree'
    return (sentiment_comparison, blob_sentiment, vader_sentiment)
```

```python
# Convert the Tweet column to string type
df_blob_vader['Tweet'] = df_blob_vader['Tweet'].astype(str)

# Apply the compare_sentiment function to each tweet in the dataframe
df_blob_vader['sentiment_comparison'], df_blob_vader['blob_sentiment'], df_blob_vader['vader_sentiment'] = zip(*df_blob_vader['Tweet'].apply(compare_sentiment))

# Print the results
print(df_blob_vader[['blob_sentiment', 'vader_sentiment', 'sentiment_comparison']])
```
